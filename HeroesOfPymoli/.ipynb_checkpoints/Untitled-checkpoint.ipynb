```python
# pd melt
# gather clumns into rows
variable_df = pd.DataFrame({"a":[1,4,7, "x"],
                  "b":[2,5,8,0],
                  "c":[3,6,9,"#"]},
                   index = [1,2,3,4])
#variable_df.head()
pd.melt(variable_df)

```
